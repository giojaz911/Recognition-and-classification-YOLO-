```python


from ultralytics import YOLO
import cv2

# 1 Cargar tu modelo entrenado
model = YOLO("best.pt")

# 2 Nombres de los animals
animals = [
    "Flamingo","Penguin"
]

# 3 Colores exactos en BGR (OpenCV usa BGR)
colors = {
    0: (0xAA, 0xAA, 0xE3),  
    1: (0xDE, 0xDC, 0xA6),
}

# 4 Abrir la webcam
cap = cv2.VideoCapture(0)  # 0 = cmara principal

if not cap.isOpened():
    print("No se pudo abrir la cmara")
    exit()

print("Presiona 'q' para salir...")

# 5 Procesar frames en tiempo real
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Detectar animals
    results = model(frame)

    # Dibujar resultados
    for result in results:
        boxes = result.boxes
        for i, box in enumerate(boxes.xyxy):
            cls_id = int(boxes.cls[i])
            conf = float(boxes.conf[i])
            x1, y1, x2, y2 = map(int, box)

            # Dibujar rectngulo y etiqueta con color exacto
            color = colors.get(cls_id, (0, 255, 0))  # verde por defecto
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            label = f"{animals[cls_id]} {conf:.2f}"
            cv2.putText(frame, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Mostrar frame
    cv2.imshow("YOLO11s - Animals", frame)

    # Salir con 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar recursos
cap.release()
cv2.destroyAllWindows()
```
