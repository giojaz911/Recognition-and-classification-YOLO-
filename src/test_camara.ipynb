```python
# Script: Detectar idols desde webcam con YOLO11n entrenado (colores exactos)
# ------------------------------------------------------------------------

from ultralytics import YOLO
import cv2

best_path = "../runs/detect/best_model/weights/best.pt"
# 1️⃣ Cargar tu modelo entrenado
model = YOLO("best.pt")

# 2️⃣ Nombres de los idols
"""idols = [
    "Chaewon","Chaeyeon","Eunbi","Hitomi","Hyewon","Minju",
    "Nako","Sakura","Wonyoung","Yena","Yujin","Yuri"
]"""

# 3️⃣ Colores exactos en BGR (OpenCV usa BGR)
colors = {
    0: (0xD5, 0xE5, 0xCE),  # Chaewon - #CEE5D5
    1: (0xDE, 0xDC, 0xA6)  # Chaeyeon - #A6DCDE
}

# 4️⃣ Abrir la webcam
cap = cv2.VideoCapture(0)  # 0 = cámara principal

if not cap.isOpened():
    print("No se pudo abrir la cámara")
    exit()

print("Presiona 'q' para salir...")

# 5️⃣ Procesar frames en tiempo real
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Detectar idols
    results = model(frame)

    # Dibujar resultados
    for result in results:
        boxes = result.boxes
        for i, box in enumerate(boxes.xyxy):
            cls_id = int(boxes.cls[i])
            conf = float(boxes.conf[i])
            x1, y1, x2, y2 = map(int, box)

            # Dibujar rectángulo y etiqueta con color exacto
            color = colors.get(cls_id, (0, 255, 0))  # verde por defecto
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            label = f"{cls_id} {conf:.2f}"
            cv2.putText(frame, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Mostrar frame
    cv2.imshow("YOLO11n - Animals", frame)

    # Salir con 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar recursos
cap.release()
cv2.destroyAllWindows()
```
